## 파생변수 추가
- 기존의 변수들을 조합하거나 함수를 사용해서 새로운 컬럼(변수)를 만드는 것
- 데이터 프레임 assign 메소드 활용

```python
exam.head()
```

```python
exam.assign(total=exam['math']+exam['english']+exam['science'])
```

```python
# 여러 파생변수 한번에 할당하기
exam.assign(
    total = exam['math'] + exam['english'] + exam['science'],
    mean = ((exam['math'] + exam['english'] + exam['science']) / 3).round(2)
).head()
```

```python
# 조건으로 파생변수 할당하기
import numpy as np
exam.assign(
    test = np.where(exam['science']>60, 'pass','fail')
).head()
```

```python
# lambda 함수 적용하기
# 데이터 프레임명 대신 익명의 x를 사용하여, 코드간결화
exam.assign(
    total = exam['math'] + exam['english'] + exam['science'],
    mena = (lambda x:x['total']/3)
    # mena = (lambda x:(x['total']/3).round(2))
).head()
```

```python
exam.assign(
    total = lambda x:x['math']+x['english']+x['science'],
    mena = lambda x:x['total']/3    
).head()
```

## 데이터합치기
- 가로로합치기 merge
- 세로로 합치기 concat

```python
test1 = pd.DataFrame({
    'id':[1,2,3,4,5],
    '중간' : [50,60,70,80,90]})
test2 = pd.DataFrame({
    'id':[1,2,3,4,5],
    '중간':[55,66,77,88,99]})
```

```python
test1
```

```python
test2
```

```python
# merge : 가로로 합치기
# how 파라미터를 통해서 병합 방식(inner,left,right) 변경
# on 파라미터로 병합할 기준을 설정
pd.merge(test1,test2,on='id')
```

```python
teacher = pd.DataFrame({
    'nclass':[1,2,3,4,5],
    'teacher': ['kim','lee','park','jung','choi']})
teacher
```

```python
# merge 를 통해 기준변수를 기준으로 새로운 컬럼 할당
pd.merge(exam, teacher,how='left',on='nclass').head()
```

```python
## 세로로 합치기 concat
test1 = pd.DataFrame({ 'id' : [1,2,3,4,5],
                      'test' : [50,60,70,80,90]})
test2 = pd.DataFrame({ 'id' : [6,7,8,9,10],
                      'test' : [55,66,77,88,99]})
```

```python
# 세로로 합칠 때는 각 데이터프레임의 컬럼명이 같아야 한다.
pd.concat([test1,test2])
```

  - c , cng, 2.35
  - d , 디젤, 2.38
  - e , 에탄올, 2.11
  - p , 프리미엄, 2.76
  - r , 레귤러, 2.22 
- 연료정보에 대한 데이터 프레임을 생성
